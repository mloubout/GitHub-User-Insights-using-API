# SlimGroup GitHub Insights

In this notebook, we look at some traffic and contribution insight for our repositories. 


## Import libraries and data

I'll import the necessary libraries as I'll use the **seaborn** library to plot the insights.

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
font = {'family' : 'normal',
        'size'   : 20}

matplotlib.rc('font', **font)
```

```python
repos = pd.read_csv('repos_info.csv')
commits = pd.read_csv('commits_info.csv')
```

## Basic Analysis

Let's begin by exploring the repos and commit datasets. Then, I'll check the repos on the basis of forks, stars and watchers. Finally, I'll check the commit messages.

```python
print("Total repos till date: {}".format(repos.shape[0]))
print("Total commits till date: {}".format(commits.shape[0]))
```

## Clones and views

```python
repo_views = repos.nlargest(10, 'Views')
plt.figure(figsize = (20, 12))
sns.barplot(repo_views['Name'], repo_views['Views'])
plt.xticks(rotation = 90)
plt.xlabel("Repository", fontsize = 20)
plt.ylabel("Viewers", fontsize = 20)
plt.title("Unique viewers per repository", fontsize = 20)
```

```python
repo_clones = repos.nlargest(10, 'Clones')

plt.figure(figsize = (20, 12))
sns.barplot(repo_clones['Name'], repo_clones['Clones'])
plt.xticks(rotation = 90)
plt.xlabel("Repository", fontsize = 20)
plt.ylabel("Clones", fontsize = 20)
plt.title("Unique clones per repository", fontsize = 20)
```

## Commits per repository

```python
commits_count = pd.DataFrame(pd.merge(repos, 
         commits, 
         left_on='Id', 
         right_on='Repo Id', 
         how = 'left').groupby('Id').size().reset_index())
commits_count.columns = ['Id', 'Commits count']
```

```python
repos = pd.merge(repos, commits_count, on = 'Id')
```

```python
plt.figure(figsize = (20, 12))
sns.barplot(repos['Name'], repos['Commits count'])
plt.xticks(rotation = 90)
plt.xlabel("Repository", fontsize = 20)
plt.ylabel("Commits", fontsize = 20)
plt.title("Commits per repository", fontsize = 20)
```

## Yearly analysis

```python
commits['Year'] = commits['Date'].apply(lambda x: x.split('-')[0])
yearly_stats = commits.groupby('Year').count()['Commit Id']
```

```python
plt.figure(figsize = (20, 12))
sns.barplot(yearly_stats.index, yearly_stats.values)
plt.xlabel("Year", fontsize = 20)
plt.ylabel("Commits", fontsize = 20)
plt.title("Commits in each year", fontsize = 20)
```

## Monthly commits in 2021

```python
commits['Month'] = commits['Date'].apply(lambda x: x.split('-')[1])
monthly_stats = commits[commits['Year'] == '2021'].groupby('Month').count()['Commit Id']
```

```python
month_mapper = { '01': 'January', '02': 'February', '03': 'March', '04': 'April', '05': 'May', '06': 'June', '07': 'July', '08': 'August', 
                '09': 'September', '10': 'October', '11': 'November', '12': 'December'}
labels = []
for value in monthly_stats.index:
    labels.append(month_mapper[value])

plt.figure(figsize = (20, 12))
plot = sns.barplot(monthly_stats.index, monthly_stats.values)
plt.xlabel("Month", fontsize = 20)
plot.set_xticklabels(labels)
plt.ylabel("Commits", fontsize = 20)
plt.title("Commits in each month of 2021", fontsize = 20)
```

## Everyday commits in October 2021


```python
commits['Date only'] = pd.to_datetime(commits['Date'].apply(lambda date: date.split('T')[0]))
commit_stats = commits[(commits['Year'] == '2021')
                       & (commits['Month'] == '10')]['Date only'].value_counts().sort_index()
```

```python
labels = []
for value in commit_stats.index:
    labels.append(value.date())
    
plt.figure(figsize = (20, 12))
plot = sns.barplot(commit_stats.index, commit_stats.values)
plt.xlabel("Date", fontsize = 20)
plt.xticks(rotation = 90)
plot.set_xticklabels(labels)
plt.ylabel("Commits", fontsize = 20)
plt.title("Commits in October, 2021", fontsize = 20)
```

## Most popular languages

Next, I'll explore languages that we've used in each repository.

```python
list_of_languages = []
for languages in repos['Languages']:
    if type(languages) == str:
        for language in languages.split(','):
            list_of_languages.append(language.strip())
```

```python
languages_count = pd.Series(list_of_languages).value_counts()

plt.figure(figsize = (20, 12))
sns.barplot(languages_count.index, languages_count.values)
plt.xlabel("Languages", fontsize = 20)
plt.xticks(rotation = 90)
plt.ylabel("Repositories", fontsize = 20)
plt.title("Language distribution amongst all repositories", fontsize = 20)
```
